# Incêndios Florestais

Os incêndios florestais são grandes queimadas que ocorrem em áreas de vegetação, como florestas, savanas e pastagens. Eles podem ser causados por fatores naturais, como raios, ou por atividades humanas, como queimadas agrícolas, desmatamento, e descarte inadequado de cigarros. Esses incêndios têm um impacto devastador no meio ambiente, destruindo a biodiversidade, emitindo grandes quantidades de gases de efeito estufa e contribuindo para as mudanças climáticas.



```python
#importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importando a base
incendios = pd.read_csv('/content/Dados_Indendio.csv', encoding='latin-1')
```

```python
#analise dos incendios por ano
analise = incendios.groupby('year')['number'].sum().reset_index()

_=plt.style.use('ggplot')
_=plt.xlabel('Anos')
_=plt.ylabel('Qnt de Queimadas')
_=plt.title('Analise de Queimadas por Ano no Brasil 1998 - 2017', fontsize=12)
_=sns.lineplot(data=analise, x='year', y='number', color='red')
_=plt.figure(figsize=(1,5))
```

```python
#meses que ocorreram mais queimadas no Brasil 1998 - 2017
analise2 = incendios.groupby('month')['number'].sum().reset_index()

_=plt.figure(figsize=(9,4))
_=plt.title('Mêses que ocorreram mais Queimadas 1998 - 2017', fontsize=12)
_=sns.barplot(data = analise2, x='month', y='number', color='green',
              order=['Janeiro','Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])
_=plt.xlabel('Mês')
_=plt.ylabel('Qnt. Queimadas')
_=plt.xticks(rotation=45, ha='right')
```

```python
#estados que ocorreram mais queimadas
analise3 = incendios.groupby('state')['number'].sum().reset_index()
ordenado = analise3.sort_values(by='number', ascending=False)

_=plt.figure(figsize=(9,4))
_=plt.title('Estados que ocorreram mais Queimadas 1998 - 2017', fontsize=12)
_=sns.barplot(data = ordenado, x='state', y='number', color='orange')
_=plt.xlabel('Estados')
_=plt.ylabel('Qnt. Queimadas')
_=plt.xticks(rotation=45, ha='right')
```
